We can categorize this column by the type of contract the players has into: free, bought and on loan 

```python
def player_status(val):
    val = val.split(" ")
    if 'Free' in val:
        return "Free"
    elif "Loan" in val:
        return "Loan"
    elif "-" in val:
        return 'Contract'
```

```python
fifa21_Data['Height_m'] = fifa21_Data.Height
fifa21_Data['Height_feet'] = fifa21_Data.Height

#NB: 1 metre = 0.3048 feet 1 inch = 1/12 feet
```

```python
def to_metres(value):
    if 'cm' in value:
        # convert from centimeters to metres
        return float(value.replace('cm', ''))/100
    elif 'm' in value:
        # height is already in metres
        return float(value.replace('m', ''))
    else:
        # height is in an unexpected format
        return value

```

```python
def to_feet(value):
    if '"' in value:
        value = value.strip('"').split("'")
        return float(value[0]) + float(value[1])/12
    else:
        # height is in an unexpected format
        return value



```

```python
fifa21_Data.Height_m = fifa21_Data.Height_m.apply(to_metres)
fifa21_Data.Height_feet = fifa21_Data.Height_feet.apply(to_feet)
```

```python
def hit_func(val):
    if "K" in str(val):
        val = val.replace('K', '')
        return int(float(val)*1000)
    else:
        return int(val)
```

```python
fifa21_Data['Hits'] = fifa21_Data['Hits'].apply(hit_func)

fifa21_Data['Hits'].unique()
```

```python
#Working on the Wage column

fifa21_Data['Wage'].unique()
```

```python
# Function will clean Column Wage
def convert_Wage(val):
    if val == '':
        return 0
    # Remove the Euro sign and any commas from the string
    val = val.replace('€','').replace(',','')
    # Check if the value is in thousands or millions
    if 'K' in val:
        return int(float(val.replace('K','')) * 1000)
    elif 'M' in val:
        return int(float(val.replace('M','')) * 1000000)
    else:
        return int(val)
     
    
```

```python
fifa21_Data['Wage']= fifa21_Data['Wage'].apply(convert_Wage)
```

```python
#Working on these columns

# define a function to remove 'star' character
def remove_star(i, column_name):
    if isinstance(i, int):
        return i
    else:
        if column_name == "Weaker Foot":
            return int(i[0])
        elif column_name == "Skill Moves":
            return int(i[0])
        elif column_name == "Injury Resistance":
            return int(i[0])


# apply the function to the columns that contain 'star' character
fifa21_Data["Weaker Foot"] = fifa21_Data["Weaker Foot"].apply(lambda x: remove_star(x, "Weaker Foot"))
fifa21_Data["Skill Moves"] = fifa21_Data["Skill Moves"].apply(lambda x: remove_star(x, "Skill Moves"))
fifa21_Data["Injury Resistance"] = fifa21_Data["Injury Resistance"].apply(lambda x: remove_star(x, "Injury Resistance"))

```

```python
def convert_RC(val):
    if isinstance(val, str):
        if val[-1] == "M":
            val = float(val[1:-1].replace(",", ".")) * 1000000
        elif val[-1] == "K":
            val = float(val[1:-1].replace(",", ".")) * 1000
        else:
            val = 0
        return val
    else:
        return val

```

```python
fifa21_Data["Release Clause"] = fifa21_Data["Release Clause"].apply(convert_RC)
```

```python
#Working on the Value column

def convert_Value(val):
    if isinstance(val, str):
        if val[-1] == "M":
            val = float(val[1:-1].replace(",", ".")) * 1000000
        elif val[-1] == "K":
            val = float(val[1:-1].replace(",", ".")) * 1000
        else:
            val = 0
        return val
    else:
        return val
```

```python
fifa21_Data["Value"] = fifa21_Data["Value"].apply(convert_Value)
```
